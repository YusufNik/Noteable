Along the way, you will learn how to connect:

- variables
- functions
- conditionals
- loops
- randomness
- repeated simulation
- data tables
- visualisation
- interactivity

By the end, you will have made a real interactive program in Python.

## 2. First Rules: Variables, Choices, and Chance

<div style="border:1px solid #c3e6cb; background:#d4edda; color:#155724; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Core idea:</b> A game is just a set of rules. In Python, those rules become variables, conditionals, and random events.
</div>

Here is the world we will build:

- You have a limited number of turns.
- Your goal is to collect enough treasure points to win.
- On each turn, you choose a path:
  - <b>Safe path</b>: smaller rewards, but more reliable
  - <b>Risky path</b>: bigger rewards, but more danger

For our first step, we will simulate just **one turn**. This is a great place to see how variables, <code>if</code> statements, and randomness fit together.

```python
print("Creating our first game turn...")

current_score = 0
target_score = 14

# EDIT THIS VALUE: try "safe" or "risky"
path_choice = "risky"

die_roll = random.randint(1, 6)

if path_choice == "safe":
    if die_roll >= 3:
        points_change = 2
    else:
        points_change = 1
elif path_choice == "risky":
    if die_roll >= 5:
        points_change = 6
    elif die_roll == 1:
        points_change = -2
    else:
        points_change = 0
else:
    points_change = 0

updated_score = max(0, current_score + points_change)

print(f"Chosen path: {path_choice}")
print(f"Die roll: {die_roll}")
print(f"Points change this turn: {points_change}")
print(f"Updated score: {updated_score}")
print(f"Target score for the full game: {target_score}")
```

## 3. Building Our Mini-Game or Simulation Functions

<div style="border:1px solid #b8daff; background:#d9edf7; color:#0c5460; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Goal:</b> Turn one-off code into reusable game functions. This is where Python starts to feel powerful: once a rule becomes a function, we can use it again and again.
</div>

We will create functions for three jobs:

- play one turn
- decide which action a strategy should choose
- play a full game from start to finish

This is also where <code>if</code> / <code>elif</code> / <code>else</code> becomes especially useful. A strategy is just a rule for making a choice.

```python
print("Creating our game functions...")

def play_one_turn(choice, current_score, rng=None):
    """Play one turn of Treasure Trail and return the result."""
    if rng is None:
        rng = random
    
    die_roll = rng.randint(1, 6)
    
    if choice == "safe":
        if die_roll >= 3:
            points_change = 2
        else:
            points_change = 1
    elif choice == "risky":
        if die_roll >= 5:
            points_change = 6
        elif die_roll == 1:
            points_change = -2
        else:
            points_change = 0
    else:
        raise ValueError("choice must be 'safe' or 'risky'")
    
    new_score = max(0, current_score + points_change)
    
    return {
        "choice": choice.title(),
        "roll": die_roll,
        "points_change": points_change,
        "score_after_turn": new_score
    }

def choose_action(strategy_name, turn_number, current_score, total_turns, target_score):
    """Choose safe or risky based on a simple strategy rule."""
    remaining_turns = total_turns - turn_number + 1
    remaining_needed = max(0, target_score - current_score)
    needed_per_turn = math.ceil(remaining_needed / remaining_turns) if remaining_turns > 0 else 0
    
    if strategy_name == "Steady Steps":
        return "safe"
    elif strategy_name == "Big Gambles":
        return "risky"
    elif strategy_name == "Pace Chaser":
        return "risky" if needed_per_turn > 2 else "safe"
    elif strategy_name == "Last Chance":
        if turn_number > total_turns - 3 and current_score < target_score:
            return "risky"
        else:
            return "safe"
    else:
        return "safe"

def play_one_game(strategy_name, total_turns=8, target_score=14, seed=None):
    """Play a full game and return both turn-by-turn history and a summary."""
    rng = random.Random(seed) if seed is not None else random
    current_score = 0
    history = []
    
    for turn_number in range(1, total_turns + 1):
        choice = choose_action(strategy_name, turn_number, current_score, total_turns, target_score)
        result = play_one_turn(choice, current_score, rng=rng)
        result["turn"] = turn_number
        current_score = result["score_after_turn"]
        history.append(result)
    
    history_df = pd.DataFrame(history)
    final_score = int(history_df["score_after_turn"].iloc[-1])
    won = final_score >= target_score
    risky_turns = int((history_df["choice"] == "Risky").sum())
    
    summary = {
        "strategy": strategy_name,
        "total_turns": total_turns,
        "target_score": target_score,
        "final_score": final_score,
        "won": won,
        "risky_turns": risky_turns
    }
    
    return history_df, summary

def simulate_many_games(strategy_name, number_of_games=500, total_turns=8, target_score=14):
    """Run many games and collect both game-level and turn-level results."""
    game_summaries = []
    turn_histories = []
    
    for game_id in range(1, number_of_games + 1):
        history_df, summary = play_one_game(
            strategy_name=strategy_name,
            total_turns=total_turns,
            target_score=target_score
        )
        summary["game_id"] = game_id
        history_df["game_id"] = game_id
        history_df["strategy"] = strategy_name
        game_summaries.append(summary)
        turn_histories.append(history_df)
    
    return pd.DataFrame(game_summaries), pd.concat(turn_histories, ignore_index=True)

print("Game functions ready!")
```

```python
print("Testing one full game with a reproducible example...")

example_history, example_summary = play_one_game(
    strategy_name="Pace Chaser",
    total_turns=8,
    target_score=14,
    seed=7
)

display(example_history[["turn", "choice", "roll", "points_change", "score_after_turn"]])
display(pd.DataFrame([example_summary]))

example_progress = pd.DataFrame({
    "turn": np.arange(1, len(example_history) + 1),
    "score": example_history["score_after_turn"]
})

fig = px.line(
    example_progress,
    x="turn",
    y="score",
    markers=True,
    template="plotly_white",
    title="One example game: score over time"
)
fig.add_hline(y=14, line_dash="dash", line_color="firebrick")
fig.update_layout(xaxis_title="Turn number", yaxis_title="Score")
fig.show()

print("One example game complete. Now we are ready to simulate many games.")
```

## 4. Running Many Rounds and Collecting Results

<div style="border:1px solid #c3e6cb; background:#d4edda; color:#155724; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Why loops matter:</b> A single game is interesting, but it does not tell us very much about a strategy. A loop lets us repeat the same process many times so we can collect evidence.
</div>

This is a core programming move:

1. define the rules once
2. repeat the simulation many times
3. store the results
4. compare what happened

```python
print("Running many games for each strategy...")

# EDIT THIS VALUE: try 200, 600, or 1000
number_of_simulations = 600

total_turns = 8
target_score = 14
strategy_names = ["Steady Steps", "Big Gambles", "Pace Chaser", "Last Chance"]

random.seed(2025)

all_game_results = []
all_turn_results = []

for strategy_name in strategy_names:
    game_df, turn_df = simulate_many_games(
        strategy_name=strategy_name,
        number_of_games=number_of_simulations,
        total_turns=total_turns,
        target_score=target_score
    )
    all_game_results.append(game_df)
    all_turn_results.append(turn_df)

combined_game_results = pd.concat(all_game_results, ignore_index=True)
combined_turn_results = pd.concat(all_turn_results, ignore_index=True)

strategy_summary = (
    combined_game_results
    .groupby("strategy")
    .agg(
        games_played=("won", "size"),
        win_rate=("won", "mean"),
        average_final_score=("final_score", "mean"),
        median_final_score=("final_score", "median"),
        best_score=("final_score", "max"),
        average_risky_turns=("risky_turns", "mean")
    )
    .reset_index()
)

strategy_summary["win_rate_pct"] = np.round(strategy_summary["win_rate"] * 100, 1)
strategy_summary["average_final_score"] = np.round(strategy_summary["average_final_score"], 2)
strategy_summary["average_risky_turns"] = np.round(strategy_summary["average_risky_turns"], 2)
strategy_summary["score_gap_to_target"] = np.round(target_score - strategy_summary["average_final_score"], 2)

display(
    strategy_summary.sort_values("win_rate_pct", ascending=False)[[
        "strategy",
        "games_played",
        "win_rate_pct",
        "average_final_score",
        "median_final_score",
        "best_score",
        "average_risky_turns",
        "score_gap_to_target"
    ]]
)

print("Simulation complete! We now have a full dataset of outcomes to visualise.")
```

## 5. Visualising Strategy and Outcomes

<div style="border:1px solid #b8daff; background:#d9edf7; color:#0c5460; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Goal:</b> Turn our simulation results into visual evidence. Charts help us see patterns that are hard to spot in raw tables.
</div>

We will look at three questions:

- Which strategy wins most often?
- How do scores grow over time?
- How spread out are the final results?

```python
print("Generating win-rate comparison chart...")

win_rate_chart = px.bar(
    strategy_summary.sort_values("win_rate_pct", ascending=False),
    x="strategy",
    y="win_rate_pct",
    color="average_final_score",
    text="win_rate_pct",
    color_continuous_scale="Viridis",
    template="plotly_white",
    title="Which strategy reaches the target most often?"
)

win_rate_chart.update_traces(texttemplate="%{text:.1f}%", textposition="outside")
win_rate_chart.update_layout(xaxis_title="Strategy", yaxis_title="Win rate (%)")
win_rate_chart.show()

print("Win-rate chart generated successfully!")
```

```python
print("Plotting average score over time...")

average_progress = (
    combined_turn_results
    .groupby(["strategy", "turn"])
    .agg(average_score=("score_after_turn", "mean"))
    .reset_index()
)

progress_chart = px.line(
    average_progress,
    x="turn",
    y="average_score",
    color="strategy",
    markers=True,
    template="plotly_white",
    title="How quickly does each strategy build score?"
)

progress_chart.add_hline(y=target_score, line_dash="dash", line_color="firebrick")
progress_chart.update_layout(xaxis_title="Turn number", yaxis_title="Average score")
progress_chart.show()

print("Progress chart ready!")
```

```python
print("Showing the spread of final scores...")

distribution_chart = px.box(
    combined_game_results,
    x="strategy",
    y="final_score",
    color="strategy",
    points="outliers",
    template="plotly_white",
    title="Final score spread across many simulated games"
)

distribution_chart.add_hline(y=target_score, line_dash="dash", line_color="firebrick")
distribution_chart.update_layout(xaxis_title="Strategy", yaxis_title="Final score", showlegend=False)
distribution_chart.show()

print("Distribution chart complete!")
```

## 6. Interactive Explorations

<div style="border:1px solid #c3e6cb; background:#d4edda; color:#155724; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Interactivity:</b> This is where the notebook becomes a small game console and simulation lab. You can play Treasure Trail yourself and then compare strategies with live controls.
</div>

Run the next cells and experiment. This is designed to feel playful, but it is also doing real programming work underneath.

```python
print("Building your playable Treasure Trail console...")

manual_game_state = {
    "turn": 1,
    "score": 0,
    "total_turns": 8,
    "target_score": 14,
    "history": []
}

manual_output = widgets.Output()
safe_button = widgets.Button(description="Take Safe Path", button_style="success")
risky_button = widgets.Button(description="Take Risky Path", button_style="warning")
reset_button = widgets.Button(description="Reset Game", button_style="info")

def render_manual_game():
    with manual_output:
        clear_output(wait=True)
        display_turn = min(manual_game_state["turn"], manual_game_state["total_turns"])
        print("Treasure Trail: play your own game")
        print(f"Turn: {display_turn} of {manual_game_state['total_turns']}")
        print(f"Current score: {manual_game_state['score']} | Target: {manual_game_state['target_score']}")
        print("Safe path = smaller steady gains. Risky path = possible big reward, but danger too.\n")
        
        if manual_game_state["history"]:
            history_df = pd.DataFrame(manual_game_state["history"])
            display(history_df[["turn", "choice", "roll", "points_change", "score_after_turn"]])
            
            progress_df = pd.DataFrame({
                "turn": [0] + history_df["turn"].tolist(),
                "score": [0] + history_df["score_after_turn"].tolist()
            })
            
            fig = px.line(
                progress_df,
                x="turn",
                y="score",
                markers=True,
                template="plotly_white",
                title="Your score so far"
            )
            fig.add_hline(y=manual_game_state["target_score"], line_dash="dash", line_color="firebrick")
            fig.update_layout(xaxis_title="Turn number", yaxis_title="Score")
            fig.show()
        else:
            print("No moves yet. Click one of the buttons to begin.")
        
        if manual_game_state["turn"] > manual_game_state["total_turns"]:
            if manual_game_state["score"] >= manual_game_state["target_score"]:
                print("Result: You reached the treasure target — you win!")
            else:
                print("Result: You fell short this time. Try a different mix of safe and risky choices.")
            safe_button.disabled = True
            risky_button.disabled = True

def make_choice_handler(choice_name):
    def handle_click(button):
        if manual_game_state["turn"] <= manual_game_state["total_turns"]:
            result = play_one_turn(choice_name, manual_game_state["score"])
            result["turn"] = manual_game_state["turn"]
            manual_game_state["score"] = result["score_after_turn"]
            manual_game_state["history"].append(result)
            manual_game_state["turn"] += 1
        render_manual_game()
    return handle_click

def reset_manual_game(button):
    manual_game_state["turn"] = 1
    manual_game_state["score"] = 0
    manual_game_state["history"] = []
    safe_button.disabled = False
    risky_button.disabled = False
    render_manual_game()

safe_button.on_click(make_choice_handler("safe"))
risky_button.on_click(make_choice_handler("risky"))
reset_button.on_click(reset_manual_game)

display(widgets.HBox([safe_button, risky_button, reset_button]))
display(manual_output)
render_manual_game()

print("Mini-game console ready! Try to reach 14 points in 8 turns.")
```

<div style="border:1px solid #ffeeba; background:#fff3cd; color:#856404; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Try it yourself:</b> Play the mini-game a few times. Do you trust the safe path more, or do you only switch to the risky path when you are behind? Your own behaviour here is a strategy too.
</div>

```python
print("Building the interactive strategy lab...")

def compare_strategies_live(strategy_a="Steady Steps", strategy_b="Pace Chaser", number_of_games=300, total_turns=8, target_score=14):
    if strategy_a == strategy_b:
        print("You chose the same strategy twice, so this comparison mainly shows random variation.")
    
    game_df_a, turn_df_a = simulate_many_games(strategy_a, number_of_games=number_of_games, total_turns=total_turns, target_score=target_score)
    game_df_b, turn_df_b = simulate_many_games(strategy_b, number_of_games=number_of_games, total_turns=total_turns, target_score=target_score)
    
    comparison_games = pd.concat([game_df_a, game_df_b], ignore_index=True)
    comparison_turns = pd.concat([turn_df_a, turn_df_b], ignore_index=True)
    
    summary_df = (
        comparison_games
        .groupby("strategy")
        .agg(
            games_played=("won", "size"),
            win_rate=("won", "mean"),
            average_final_score=("final_score", "mean"),
            best_score=("final_score", "max")
        )
        .reset_index()
    )
    
    summary_df["win_rate_pct"] = (summary_df["win_rate"] * 100).round(1)
    summary_df["average_final_score"] = summary_df["average_final_score"].round(2)
    display(summary_df[["strategy", "games_played", "win_rate_pct", "average_final_score", "best_score"]])
    
    comparison_bar = px.bar(
        summary_df,
        x="strategy",
        y="win_rate_pct",
        color="average_final_score",
        text="win_rate_pct",
        color_continuous_scale="Plasma",
        template="plotly_white",
        title="Live strategy comparison"
    )
    comparison_bar.update_traces(texttemplate="%{text:.1f}%", textposition="outside")
    comparison_bar.update_layout(xaxis_title="Strategy", yaxis_title="Win rate (%)")
    comparison_bar.show()
    
    live_progress = (
        comparison_turns
        .groupby(["strategy", "turn"])
        .agg(average_score=("score_after_turn", "mean"))
        .reset_index()
    )
    
    progress_line = px.line(
        live_progress,
        x="turn",
        y="average_score",
        color="strategy",
        markers=True,
        template="plotly_white",
        title="Average score by turn in the live simulation"
    )
    progress_line.add_hline(y=target_score, line_dash="dash", line_color="firebrick")
    progress_line.update_layout(xaxis_title="Turn number", yaxis_title="Average score")
    progress_line.show()
    
    print(f"Simulated {number_of_games} games for each strategy.")
    print("Tip: increase the number of games to make the comparison more stable.")

strategy_options = ["Steady Steps", "Big Gambles", "Pace Chaser", "Last Chance"]

strategy_a_dropdown = widgets.Dropdown(
    options=strategy_options,
    value="Steady Steps",
    description="Strategy A:"
)

strategy_b_dropdown = widgets.Dropdown(
    options=strategy_options,
    value="Pace Chaser",
    description="Strategy B:"
)

simulation_slider = widgets.IntSlider(
    value=300,
    min=100,
    max=1500,
    step=100,
    description="Games:",
    continuous_update=False
)

turn_slider = widgets.IntSlider(
    value=8,
    min=6,
    max=12,
    step=1,
    description="Turns:",
    continuous_update=False
)

target_slider = widgets.IntSlider(
    value=14,
    min=10,
    max=22,
    step=1,
    description="Target:",
    continuous_update=False
)

print("Widget ready! Compare strategies live below:\n")
widgets.interact(
    compare_strategies_live,
    strategy_a=strategy_a_dropdown,
    strategy_b=strategy_b_dropdown,
    number_of_games=simulation_slider,
    total_turns=turn_slider,
    target_score=target_slider
);
```

## 7. What We Learned, Limits, and Next Steps

<div style="border:1px solid #c3e6cb; background:#d4edda; color:#155724; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Important:</b> Simulation is powerful, but it does not remove uncertainty. It helps us study a simplified world more carefully.
</div>

A few good interpretation habits:

- <b>Random outcomes vary.</b> One game can be lucky or unlucky.
- <b>One run is not enough.</b> That is why loops and repeated simulations matter.
- <b>A strategy can look good in one setup and weaker in another.</b> Changing the number of turns or the target score can change the results.
- <b>This game is a model, not reality.</b> Our rules are simplified on purpose so we can focus on programming ideas clearly.

Even with those limits, the notebook teaches something important: once you can encode rules, Python lets you explore “what if?” questions in a structured way.
