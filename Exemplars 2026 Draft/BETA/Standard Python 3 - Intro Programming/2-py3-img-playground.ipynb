```python
print("Starting setup: importing libraries...")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import ipywidgets as widgets

from IPython.display import display, clear_output
from skimage import data, color, exposure, filters, util

print("Success! All libraries imported. Our image playground is ready.")
```

## 2. First Steps: Images as Data

<div style="border:1px solid #c3e6cb; background:#d4edda; color:#155724; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Big idea:</b> An image is not mysterious to Python. It is a grid of numbers. For colour images, each pixel usually stores three values: red, green, and blue.
</div>

Let's begin with a small gallery of built-in sample images. These are bundled with scikit-image, so they are reliable and do not need any external download.

```python
print("Loading built-in sample images...")

sample_images = {
    "Astronaut": data.astronaut(),
    "Coffee": data.coffee(),
    "Camera": data.camera()
}

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, (name, image) in zip(axes, sample_images.items()):
    cmap = "gray" if image.ndim == 2 else None
    ax.imshow(image, cmap=cmap)
    ax.set_title(name, fontsize=13)
    ax.axis("off")

plt.tight_layout()
plt.show()

gallery_summary = []
for name, image in sample_images.items():
    gallery_summary.append({
        "Image": name,
        "Height": image.shape[0],
        "Width": image.shape[1],
        "Channels": 1 if image.ndim == 2 else image.shape[2],
        "Data type": str(image.dtype)
    })

print("Sample image gallery ready! Here is a quick summary:")
display(pd.DataFrame(gallery_summary))
```

Now let's inspect one image more closely.

When we ask Python for an image's <code>shape</code>, we are asking:

- how tall is it?
- how wide is it?
- how many colour channels does it contain?

That might sound technical, but it is really just a structured way of describing what the computer is looking at.

```python
print("Inspecting one image as data...")

# EDIT THIS VALUE: try "Coffee" or "Camera"
selected_image_name = "Astronaut"

selected_image = sample_images[selected_image_name]
height, width = selected_image.shape[:2]
channels = 1 if selected_image.ndim == 2 else selected_image.shape[2]

# EDIT THESE VALUES: choose a safe pixel position within the image
row_index = 150
column_index = 250
pixel_value = selected_image[row_index, column_index]

summary_table = pd.DataFrame([
    {
        "Selected image": selected_image_name,
        "Height": height,
        "Width": width,
        "Channels": channels,
        "Shape": str(selected_image.shape),
        "Pixel value at chosen point": str(pixel_value)
    }
])

display(summary_table)

plt.figure(figsize=(6, 6))
plt.imshow(selected_image, cmap="gray" if selected_image.ndim == 2 else None)
plt.scatter([column_index], [row_index], c="lime", s=120, marker="x")
plt.title(f"{selected_image_name}: a single pixel highlighted")
plt.axis("off")
plt.show()

print(f"The image shape is {selected_image.shape}.")
print(f"The highlighted pixel at row {row_index}, column {column_index} has value: {pixel_value}")
```

## 3. Building Our First Image Functions

<div style="border:1px solid #b8daff; background:#d9edf7; color:#0c5460; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Goal:</b> Create reusable functions that turn one-off experiments into a small toolkit. This is one of the most useful programming habits you can build early on.
</div>

A function is like a mini machine: it takes an input, does something predictable, and gives back an output. For image work, that is perfect because we often want to apply the same transformation many times.

```python
print("Creating reusable image functions...")

def to_float_image(image):
    """Convert an image to floating-point format between 0 and 1."""
    return util.img_as_float(image)

def to_grayscale(image):
    """Convert a colour image to grayscale. Leave grayscale images as they are."""
    if image.ndim == 3:
        return color.rgb2gray(image)
    return to_float_image(image)

def crop_center(image, crop_fraction=0.7):
    """Crop the central part of an image using a fraction of the original size."""
    height, width = image.shape[:2]
    new_height = int(height * crop_fraction)
    new_width = int(width * crop_fraction)
    start_row = (height - new_height) // 2
    start_col = (width - new_width) // 2
    return image[start_row:start_row + new_height, start_col:start_col + new_width]

def blur_image(image, sigma=2.0):
    """Smooth an image. Higher sigma means stronger blur."""
    channel_axis = -1 if image.ndim == 3 else None
    return filters.gaussian(image, sigma=sigma, channel_axis=channel_axis)

def edge_image(image, sigma=1.5):
    """Highlight major edges after a little smoothing."""
    gray = to_grayscale(image)
    smooth = filters.gaussian(gray, sigma=sigma)
    return filters.sobel(smooth)

def threshold_image(image, threshold=0.5):
    """Turn grayscale intensities into a black-and-white mask."""
    gray = to_grayscale(image)
    return gray > threshold

def adjust_contrast(image, gamma=1.0):
    """Adjust contrast/brightness with gamma correction."""
    return exposure.adjust_gamma(to_float_image(image), gamma=gamma)

def show_comparison(original, transformed, original_title="Original", transformed_title="Transformed"):
    """Display two images side by side."""
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    axes[0].imshow(original, cmap="gray" if original.ndim == 2 else None)
    axes[0].set_title(original_title)
    axes[0].axis("off")
    
    transformed_cmap = "gray" if getattr(transformed, "ndim", 2) == 2 else None
    axes[1].imshow(transformed, cmap=transformed_cmap)
    axes[1].set_title(transformed_title)
    axes[1].axis("off")
    
    plt.tight_layout()
    plt.show()

# EDIT THIS VALUE: try 0.5, 0.7, or 0.85
crop_fraction = 0.7
cropped_image = crop_center(selected_image, crop_fraction=crop_fraction)

show_comparison(
    selected_image,
    cropped_image,
    original_title=f"Original {selected_image_name}",
    transformed_title=f"Centre crop ({int(crop_fraction * 100)}%)"
)

print("Function toolkit ready! We have already used one of them to create a crop.")
```

    <b>Why this matters:</b> Once a function exists, you can use it again and again with different images and different parameter values. That is one of the clearest ways programming saves time and supports experimentation.
</div>

## 4. Transforming and Comparing Images

<div style="border:1px solid #b8daff; background:#d9edf7; color:#0c5460; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Goal:</b> Apply several transformations and compare the results visually. This is the heart of computational thinking here: change a parameter, inspect the result, and decide what it means.
</div>

We will work through a few different kinds of transformations. Notice how each one connects directly to a programming idea:

- a function call
- an input image
- one or more adjustable parameters
- a visible output

```python
print("Comparing grayscale conversion and two blur strengths...")

gray_image = to_grayscale(selected_image)
soft_blur = blur_image(selected_image, sigma=1.0)
strong_blur = blur_image(selected_image, sigma=4.0)

fig, axes = plt.subplots(1, 4, figsize=(18, 5))

axes[0].imshow(selected_image, cmap="gray" if selected_image.ndim == 2 else None)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(gray_image, cmap="gray")
axes[1].set_title("Grayscale")
axes[1].axis("off")

axes[2].imshow(soft_blur, cmap="gray" if soft_blur.ndim == 2 else None)
axes[2].set_title("Blur: sigma = 1.0")
axes[2].axis("off")

axes[3].imshow(strong_blur, cmap="gray" if strong_blur.ndim == 2 else None)
axes[3].set_title("Blur: sigma = 4.0")
axes[3].axis("off")

plt.tight_layout()
plt.show()

print("Comparison complete: stronger blur smooths away more detail.")
```

Blur is a great first example of a parameter at work. The <code>sigma</code> value controls the strength of the effect.

- lower sigma keeps more detail
- higher sigma gives a smoother, softer image

This is a classic programming pattern: the same function behaves differently because one input value changes.

```python
print("Exploring edges and thresholding...")

camera_image = sample_images["Camera"]
camera_gray = to_grayscale(camera_image)
camera_edges = edge_image(camera_image, sigma=1.5)
low_threshold = threshold_image(camera_image, threshold=0.35)
high_threshold = threshold_image(camera_image, threshold=0.60)

fig, axes = plt.subplots(1, 4, figsize=(18, 5))

axes[0].imshow(camera_gray, cmap="gray")
axes[0].set_title("Original grayscale")
axes[0].axis("off")

axes[1].imshow(camera_edges, cmap="magma")
axes[1].set_title("Edges")
axes[1].axis("off")

axes[2].imshow(low_threshold, cmap="gray")
axes[2].set_title("Threshold = 0.35")
axes[2].axis("off")

axes[3].imshow(high_threshold, cmap="gray")
axes[3].set_title("Threshold = 0.60")
axes[3].axis("off")

plt.tight_layout()
plt.show()

print("Edges highlight boundaries. Thresholding turns shades into a simpler black-and-white decision.")
```

<div style="border:1px solid #ffeeba; background:#fff3cd; color:#856404; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Creative exploration:</b> Go back to the cell above and change the threshold values. Try something like <code>0.20</code> and <code>0.75</code>. Which details disappear? Which details remain?
</div>

```python
print("Comparing contrast settings and plotting intensity histograms...")

lighter_gamma = adjust_contrast(camera_image, gamma=0.7)
darker_gamma = adjust_contrast(camera_image, gamma=1.4)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(camera_gray, cmap="gray")
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(lighter_gamma, cmap="gray")
axes[1].set_title("Gamma = 0.7")
axes[1].axis("off")

axes[2].imshow(darker_gamma, cmap="gray")
axes[2].set_title("Gamma = 1.4")
axes[2].axis("off")

plt.tight_layout()
plt.show()

sample_step = 4
hist_df = pd.DataFrame({
    "intensity": np.concatenate([
        camera_gray[::sample_step, ::sample_step].ravel(),
        lighter_gamma[::sample_step, ::sample_step].ravel(),
        darker_gamma[::sample_step, ::sample_step].ravel()
    ]),
    "version": (["Original"] * camera_gray[::sample_step, ::sample_step].size) +
               (["Gamma 0.7"] * lighter_gamma[::sample_step, ::sample_step].size) +
               (["Gamma 1.4"] * darker_gamma[::sample_step, ::sample_step].size)
})

histogram_figure = px.histogram(
    hist_df,
    x="intensity",
    color="version",
    nbins=40,
    opacity=0.65,
    barmode="overlay",
    template="plotly_white",
    title="Pixel intensity distributions under different gamma settings"
)

histogram_figure.update_layout(xaxis_title="Pixel intensity", yaxis_title="Count")
histogram_figure.show()

print("Histogram generated successfully! It shows how changing gamma redistributes brightness levels.")
```

## 5. Interactive Explorations

<div style="border:1px solid #c3e6cb; background:#d4edda; color:#155724; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Interactivity:</b> Static code is useful, but interactive controls make programming feel alive. In the mini-lab below, you can change the image, choose a transformation, and adjust parameters live.
</div>

Run the next cell, then explore. This is a great moment to experiment: you can safely change settings and immediately see what happens.

```python
print("Building the interactive image lab...")

def apply_transformation(image, transform_name, sigma=2.0, threshold=0.5, gamma=1.0, crop_fraction=0.7):
    if transform_name == "Original":
        return to_float_image(image), "Original"
    elif transform_name == "Grayscale":
        return to_grayscale(image), "Grayscale"
    elif transform_name == "Blur":
        return blur_image(image, sigma=sigma), f"Blur (sigma = {sigma})"
    elif transform_name == "Edges":
        return edge_image(image, sigma=sigma), f"Edges (sigma = {sigma})"
    elif transform_name == "Threshold":
        return threshold_image(image, threshold=threshold), f"Threshold ({threshold:.2f})"
    elif transform_name == "Contrast":
        return adjust_contrast(image, gamma=gamma), f"Gamma = {gamma:.2f}"
    elif transform_name == "Crop":
        return crop_center(image, crop_fraction=crop_fraction), f"Centre crop ({int(crop_fraction * 100)}%)"
    else:
        return to_float_image(image), "Original"

def interactive_image_lab(image_name="Astronaut", transform_name="Blur", sigma=2.0, threshold=0.5, gamma=1.0, crop_fraction=0.7):
    image = sample_images[image_name]
    transformed, transformed_title = apply_transformation(
        image,
        transform_name=transform_name,
        sigma=sigma,
        threshold=threshold,
        gamma=gamma,
        crop_fraction=crop_fraction
    )
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    axes[0].imshow(image, cmap="gray" if image.ndim == 2 else None)
    axes[0].set_title(f"Original: {image_name}")
    axes[0].axis("off")
    
    transformed_cmap = "gray" if getattr(transformed, "ndim", 2) == 2 else None
    axes[1].imshow(transformed, cmap=transformed_cmap)
    axes[1].set_title(transformed_title)
    axes[1].axis("off")
    
    plt.tight_layout()
    plt.show()
    
    height, width = image.shape[:2]
    channels = 1 if image.ndim == 2 else image.shape[2]
    print(f"Image shape: {image.shape} | Height: {height} | Width: {width} | Channels: {channels}")
    print(f"Current transformation: {transform_name}")
    
    if transform_name == "Blur":
        print("Tip: higher sigma makes the image smoother and less detailed.")
    elif transform_name == "Edges":
        print("Tip: edge detection looks for strong changes in brightness.")
    elif transform_name == "Threshold":
        print("Tip: thresholding turns each pixel into a yes/no decision based on brightness.")
    elif transform_name == "Contrast":
        print("Tip: gamma changes how brightness values are distributed across the image.")
    elif transform_name == "Crop":
        print("Tip: cropping focuses attention on a smaller region of the image.")

image_dropdown = widgets.Dropdown(
    options=sorted(sample_images.keys()),
    value="Astronaut",
    description="Image:"
)

transform_dropdown = widgets.Dropdown(
    options=["Original", "Grayscale", "Blur", "Edges", "Threshold", "Contrast", "Crop"],
    value="Blur",
    description="Transform:"
)

sigma_slider = widgets.FloatSlider(
    value=2.0,
    min=0.5,
    max=6.0,
    step=0.5,
    description="Sigma:",
    continuous_update=False
)

threshold_slider = widgets.FloatSlider(
    value=0.5,
    min=0.1,
    max=0.9,
    step=0.05,
    description="Threshold:",
    continuous_update=False
)

gamma_slider = widgets.FloatSlider(
    value=1.0,
    min=0.4,
    max=2.0,
    step=0.1,
    description="Gamma:",
    continuous_update=False
)

crop_slider = widgets.FloatSlider(
    value=0.7,
    min=0.3,
    max=1.0,
    step=0.05,
    description="Crop size:",
    continuous_update=False
)

print("Widget ready! Build your own image transformations below:\n")

widgets.interact(
    interactive_image_lab,
    image_name=image_dropdown,
    transform_name=transform_dropdown,
    sigma=sigma_slider,
    threshold=threshold_slider,
    gamma=gamma_slider,
    crop_fraction=crop_slider
);
```

## 6. What We Learned, Limits, and Next Steps

<div style="border:1px solid #c3e6cb; background:#d4edda; color:#155724; padding:16px; margin:8px 0; border-radius:4px;">
    <b>Important:</b> These transformations are useful and visually powerful, but they are still simple image-processing operations. They are not the same as deep visual understanding or human interpretation.
</div>

Here are some responsible ways to interpret what we have done:

- <b>Filters are not intelligence.</b> Blurring, thresholding, and edge detection apply rules to pixel data. They do not "understand" the image.
- <b>Parameter choices matter.</b> A different sigma, threshold, or gamma value can change the result a lot.
- <b>Processed images are interpretations.</b> They highlight some features and suppress others.
- <b>Sample images are simplified demonstrations.</b> Real-world image data can be larger, noisier, and harder to interpret.

That said, the notebook still teaches something big and real: computational thinking often begins with a simple question such as, "What happens if I change this input?"
